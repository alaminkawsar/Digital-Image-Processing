```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
img_path = './Assignment/1810576130_Assignnment3/flower.jpeg'
img = plt.imread(img_path)
img = np.array(img)

def plot_img(x, y, pos, img, img_title):
    plt.subplot(x,y,pos)
    plt.title(img_title)
    plt.imshow(img,cmap='gray')
    
    # plt.subplot(x,y,pos+1)
    # plt.title('RGB Histogram')
    # plt.hist(img.ravel(),256,[0,256])



def main():
    x, y = 3, 2
    pos, channel = 1, 0

    #define six kernal
    Vertical_kernel = np.array([[0,1,0],[1,-4,1],[0,1,0]]) #kernal size 3x3
    Sobel_vertical_kernel = np.array([[-1,0,1],[-2,0,2],[-1,0,1]]) #kernal size 3x3

    #Neighborhood Processing
    '''	Convert the RGB image into grayscale and binary image.	'''
    grayscale = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
    Vertical_kernel_output = cv.filter2D(grayscale, -1, Vertical_kernel)
    Sobel_kernel_output = cv.filter2D(grayscale, -1, Sobel_vertical_kernel)

    plot_img(x,y,pos,img,"RGB Color Image")
    plot_img(x,y,pos+1,grayscale,"Gray Scale Image")
    plot_img(x,y,pos+2,Vertical_kernel_output,"Laplacian Kernel Effect output")
    plot_img(x,y,pos+3,Sobel_kernel_output,"Sobel_vertical_kernel output")
    
    plt.subplots_adjust(left=0.1,
                    bottom=0.01, 
                    right=1, 
                    top=0.95, 
                    wspace=0.0, 
                    hspace=0.7)
    
    plt.savefig("convolution filters.jpeg")


if __name__ == '__main__':
    main()
```
