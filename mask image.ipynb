```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def main():
    rgb=plt.imread("./Assignment/1810576130_Assignnment3/flower.jpeg")
    plt.subplot(2,2,1)
    plt.title("rgb")
    plt.imshow(rgb)

    grayscale=cv2.cvtColor(rgb,cv2.COLOR_RGB2GRAY)
    plt.subplot(2,2,2)
    plt.title("grayscale")
    plt.imshow(grayscale,cmap="gray")

    zero_mask=np.zeros(grayscale.shape ,dtype=np.uint8)
    zero_mask=cv2.circle(zero_mask,(950,550),300,(255,255,255),-1)

    mask_img=cv2.bitwise_and(grayscale,zero_mask)

    plt.subplot(2,2,3)
    plt.title("mask_img")
    plt.imshow(mask_img,cmap='gray')

    plt.savefig("masking.jpeg")
if __name__=='__main__':
    main()
```
